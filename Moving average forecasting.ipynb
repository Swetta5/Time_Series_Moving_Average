```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample time-series data (Monthly sales of a store)
data = {'Date': ['2024-01', '2024-02', '2024-03', '2024-04', '2024-05', '2024-06'],
        'Sales': [100, 120, 130, 110, 150, 170]}

# Create DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Convert to datetime format
df.set_index('Date', inplace=True)  # Set Date as index

# Calculate 3-month moving average
df['Forecast'] = df['Sales'].rolling(window=3).mean()

# Plot time-series data
df.plot(marker='o', title="Basic Time-Series Trend & Forecast", figsize=(6,4))
plt.xlabel("Time")
plt.ylabel("Sales")
plt.grid()
plt.show()
```
